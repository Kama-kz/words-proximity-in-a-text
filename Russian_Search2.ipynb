```
f = open('textbook.txt', encoding="utf8", errors='ignore')
#ff = f.decode("utf-8").split("\r\n")

text = f.read()
text
```

```
import nltk
from nltk import word_tokenize
from nltk import sent_tokenize
nltk.download('punkt')
from nltk.stem.snowball import SnowballStemmer 
```

```
#Весь текст разделяем на слова. ВСЕ делится на слова. Разделитель пробел.
tokens = word_tokenize(text)
print(tokens)
```

```
#создали пустой словарь для слов
dic=[[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]]
```

```
for x in tokens:
  if not x in dic[len(x)]:
    dic[len(x)].append(x)
```

вот наш словарь всех слов внизу. Разбит по кол-ву знаков в слове. 

```
for x in range(len(dic)):
   print(dic[x])
```

```
code=[]
for x in tokens:
  if x in dic[len(x)]:
    code.append('%02d%d' % ((len(x)), (dic[len(x)].index(x)+1)))
  else:
    code.append("Empty")
```

теперь кодируем текст книги в коды слов

```
print(code)
```

```
#кол-во всех слов в книге
len(code)
```

```
#кол-во уникальных слов в книге
voc={}
codeset = set(code)
len(codeset)
```

```
for item in codeset:
  voc.update({item: [i for i, x in enumerate(code) if x == item]})
```

ниже список каждлго слова в книге и где в книге они возникали. Идет номер слова напрмиер 09249 - это слово и : показывает место в книге, ну и соот-вно видно как много раз оно есть в книге

```
#input your words
#тут пиши 2 слова которые хотим проверить как близко в тексте они попажаются. все варианты. любые слова можно брать. я взял improve, code
word1="клеточный"
word2="вирус"
```

```
#берем корень согласно пакету резки слов слова №1
stemmer = SnowballStemmer("russian") 
s1=stemmer.stem(word1)
s1
```

```
#берем корень согласно пакету резки слов слова №2
s2=stemmer.stem(word2)
s2
```

```
#все слова в книге - берется корень каждого слова в книге и проверяется совпадение с нашим корнем слова №1 если совпало, то в словарь заносится номер где это слово стоит в книге
swords=[]
l=[stemmer.stem(word) for word in tokens]
for i, j in enumerate(l):
  if j == s1:
    swords.append(i)
```

```
#все слова в книге - берется корень каждого слова в книге и проверяется совпадение с нашим корнем слова №2 если совпало, то в словарь заносится номер где это слово стоит в книге
swords2=[]
l=[stemmer.stem(word) for word in tokens]
for i, j in enumerate(l):
  if j == s2:
    swords2.append(i)
```

```
#номера в книше где стоят эти слова с одним корнем, но разные по написанию
swords
```

```
#номера в книше где стоят эти слова с одним корнем, но разные по написанию - тоже самое для слова №2
swords2
```

```
#печатаем что это за слова в книге по слову №1
similar=[]
for i in swords:
  similar.append(tokens[i])
similar
```

```

#печатаем что это за слова в книге по слову №2
similar2=[]
for i in swords2:
  similar2.append(tokens[i])
similar2
```

```
#печатаем уникальные слова в книге по слову №1

usimilar=set(similar)
usimilar
```

```
#печатаем уникальные слова в книге по слову №2

usimilar2=set(similar2)
usimilar2
```

```
#берем их индексы в словаре по слову №1

indexes1=[]
for i in usimilar:
   indexes1.append('%02d%d' % ((len(i)), (dic[len(i)].index(i)+1)))
indexes1   
```

```

#берем их индексы в словаре по слову №2
indexes2=[]
for i in usimilar2:
   indexes2.append('%02d%d' % ((len(i)), (dic[len(i)].index(i)+1)))
indexes2  
```

```
#номер слова 1 в нашем словаре
word1index=('%02d%d' % ((len(word1)), (dic[len(word1)].index(word1)+1)))
word1index
```

```
#номер слова 2 в нашем словаре

word2index=('%02d%d' % ((len(word2)), (dic[len(word2)].index(word2)+1)))
word2index
```

```
#частота слова 1 в книге
word1freq=voc.get(word1index)
word1freq
```

```
#cтолько раз слово 1 повторялось в книше
len(word1freq)
```

```
#берем все места в книше где попадался корень по слову №1

all1=[]
for i in indexes1:
  all1.append(voc.get(i))
all1=[item for sublist in all1 for item in sublist]
all1
```

```
#частота слова 2 в книге

word2freq=voc.get(word2index)
word2freq
```

```
#cтолько раз слово 2 повторялось в книше

len(word2freq)
```

```
#берем все места в книше где попадался корень по слову №2

all2=[]
for i in indexes2:
  all2.append(voc.get(i))
all2=[item for sublist in all2 for item in sublist]
all2
```

```
#ищем когда эти 2 слова встречалтсь в книше ближе всего. Ставим параметр 100. можно и ближе. 100 - это 100 слов между ними в разные стороны.
distance={}
for n in all1:
   for m in all2:
     d=abs(n-m)
     if d <100:
       distance.update({n:d})
  
```

```
#вот список когда они были в тексте ближе всего др к др
distance
distancesorted = sorted(distance.items(), key=lambda x: x[1])
distancesorted
```

```
n1=distancesorted[0]
n11=n1[0]
n22=n1[1]
n33=n11+n22+1
n44=n11-n22
n11
```

```
#ну и финал. Печатаем текст где эти 2 слова попадались ближе всего др к др
print(tokens[n44:n33])
```
