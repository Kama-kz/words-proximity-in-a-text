```
import nltk
from nltk import word_tokenize
from nltk import sent_tokenize
nltk.download('punkt')
from nltk.stem.snowball import SnowballStemmer 
```

```
#    .    .  .
tokens = word_tokenize(text)
print(tokens)
```

```
#    
dic=[[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]]
```

```
for x in tokens:
  if not x in dic[len(x)]:
    dic[len(x)].append(x)
```

```
code=[]
for x in tokens:
  if x in dic[len(x)]:
    code.append('%02d%d' % ((len(x)), (dic[len(x)].index(x)+1)))
  else:
    code.append("Empty")
```

```
print(code)
```

```
#-    
voc={}
codeset = set(code)
len(codeset)
```

```
for item in codeset:
  voc.update({item: [i for i, x in enumerate(code) if x == item]})
```
